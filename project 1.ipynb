```python
#for data cleaning 
import pandas as pd
#for data in format
import numpy as np
#for dat visiulization
import plotly.express as px
import plotly.graph_ojects as go
```

```python
#`1-st step:-objective
data=pd.read_csv("apple_products.csv")
```

```python
data
```

```python
# 2nd step 
print(data.isnull().sum()) #this is for chechking any value is null or not in sheet
```

```python
#descriptive analysis means max,min,mean prices etc.
print(data.describe())
```

```python
#top 10 apple phones that sells in india :- accordin to rating-4.8,4.6
highest_rated=data.sort_values(by=["Star Rating"],ascending=False)
highest_rateed=highest_rated.head(10)
print(highest_rated['Product Name'])
```

#lets have a look at the number of rating of the highest rated iphone on flipkart

```python
iphones = highest_rated['Product Name'].value_counts()  # Note the 's' at the end of value_counts
```

```python
iphones
```

```python
iphones = highest_rated['Product Name'].value_counts() 
labels=iphones.index
counts=highest_rated['Number Of Ratings']
#for graph 
figure=px.bar(highest_rated,x=labels,y=counts,title="Number of highest rated iphones")
figure.show()
```

```python
#graphs according to reviews
#graphs according to reviews
iphone=highest_rated['Product Name'].value_counts()  # Changed from value.counts() to value_counts()
labels=iphone.index
counts=highest_rated['Number Of Reviews']
figure=px.bar(highest_rated, x=labels, y=counts, title='Number Of Reviews')  # Added comma between x=labels and y=counts
figure.show()
```

```python
#in bubble graph - y axis:rating ,x axis:price
figure = px.scatter(data_frame=data,
                   x='Number Of Ratings',  # Corrected column name with 's' at the end
                   y='Sale Price',
                   size='Discount Percentage',
                   trendline='ols',
                   title='relationship between sale price and number of ratings')
figure.show()
```
